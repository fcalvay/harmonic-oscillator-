```python
from scipy.integrate import solve_ivp
import numpy as np
import matplotlib.pyplot as plt

# Constants
A = 1       # Amplitude
omega = 2 * np.pi  # Angular frequency, setting it for 1 complete cycle per second
phi = 0     # Phase
t = np.linspace(0, 2, 1000)  # Time from 0 to 2 seconds, 1000 points

# Harmonic oscillator equations
x = A * np.cos(omega * t + phi)  # Position as a function of time
p = -A * omega * np.sin(omega * t + phi)  # Momentum as a function of time, m=1 for simplicity

# Plotting the phase space
plt.figure(figsize=(8, 6))
plt.plot(x, p)
plt.title('Phase Space of a Classical Harmonic Oscillator')
plt.xlabel('Position (x)')
plt.ylabel('Momentum (p)')
plt.grid(True)
plt.show()




# Constants
omega = 2 * np.pi # Angular frequency
m = 1  # Mass (kg)

# Equations of motion for the system
def equations_of_motion(t, y):
    x, v = y
    dxdt = v
    dvdt = -omega**2 * x
    return [dxdt, dvdt]

# Initial conditions: x(0) = A, v(0) = 0 (maximum displacement, no initial velocity)
initial_conditions = [A, 0]

# Time span
t_span = [0, 2]  # From 0 to 2 seconds
t_eval = np.linspace(*t_span, 1000)  # Time points where the solution is computed

# Solve the equations of motion
solution = solve_ivp(equations_of_motion, t_span, initial_conditions, t_eval=t_eval, method='RK45')

# Extracting position and velocity from the solution
x_solution = solution.y[0]
v_solution = solution.y[1]
p_solution = m * v_solution  # p = mv

# Plotting the phase space
plt.figure(figsize=(8, 6))
plt.plot(x_solution, p_solution)
plt.title('Phase Space of a Classical Harmonic Oscillator (Numerical Solution)')
plt.xlabel('Position (x)')
plt.ylabel('Momentum (p)')
plt.grid(True)
plt.show()
```
